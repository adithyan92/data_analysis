```python
import pandas as pd
import numpy as np
```

```python
analyst_tasksheet = pd.read_csv(r"C:\Users\Navya.LAPTOP-AQ18NGM8\Downloads\Aug-Feb.csv")
analyst_tasksheet.shape
```

```python
analyst_tasksheet["Master Station"].unique()
```

Grouping PA station time for each individual Case ID

```python
analyst_PA = analyst_tasksheet[analyst_tasksheet["Master Station"] =='PA']
analyst_PA.head()

analyst_PA = analyst_PA.groupby(["Case ID","Master Station"]).sum().rename(columns = {"Actual Duration (in min)": "PA duration"}).reset_index()
analyst_PA = analyst_PA[["Case ID","PA duration"]]
analyst_PA.head()


#analyst_PA.columns
#columns = ['Case ID', 'Master Station', 'Child Station',
#       'Actual Duration (in min)']
```

Grouping IX_SRV station time for each individual Case ID

```python
analyst_IX = analyst_tasksheet[analyst_tasksheet["Master Station"] =='IX_SRV']
analyst_IX.head()

analyst_IX = analyst_IX.groupby(["Case ID","Master Station"]).sum().rename(columns = {"Actual Duration (in min)": "IX duration"}).reset_index()
analyst_IX = analyst_IX[["Case ID","IX duration"]]
analyst_IX.head()
```

Grouping PR station time for each individual Case ID

```python
analyst_pr = analyst_tasksheet[analyst_tasksheet["Master Station"] =='PR']

analyst_pr.head()

analyst_pr = analyst_pr.groupby(["Case ID","Master Station"]).sum().rename(columns = {"Actual Duration (in min)": "PR duration"}).reset_index()
analyst_pr = analyst_pr[["Case ID","PR duration"]]
analyst_pr.head()
```

Grouping XR station time for each individual Case ID

```python
analyst_XR = analyst_tasksheet[analyst_tasksheet["Master Station"] =='XR']
analyst_XR.head()

analyst_XR = analyst_XR.groupby(["Case ID","Master Station"]).sum().rename(columns = {"Actual Duration (in min)": "XR duration"}).reset_index()
analyst_XR = analyst_XR[["Case ID","XR duration"]]
analyst_XR.head()
```

Grouping APPROVAL station time for each individual Case ID

```python
analyst_app = analyst_tasksheet[analyst_tasksheet["Master Station"] =='APPROVAL']
analyst_app.head()

analyst_app = analyst_app.groupby(["Case ID","Master Station"]).sum().rename(columns = {"Actual Duration (in min)": "Approval duration"}).reset_index()
analyst_app = analyst_app[["Case ID","Approval duration"]]
analyst_app.head()
```

Grouping Followup station time for each individual Case ID

```python
analyst_flup = analyst_tasksheet[analyst_tasksheet["Master Station"] =='FLUP']
analyst_flup.head()

analyst_flup = analyst_flup.groupby(["Case ID","Master Station"]).sum().rename(columns = {"Actual Duration (in min)": "FLUP duration"}).reset_index()
analyst_flup = analyst_flup[["Case ID","FLUP duration"]]
analyst_flup.head()
```

Merging these created station times into a single report not excluding any Case ID

```python
analyst_total = pd.merge(analyst_PA,analyst_IX,how="outer",on="Case ID")
analyst_total
```

```python
analyst_total = pd.merge(analyst_total,analyst_pr,how = "outer",on="Case ID")
analyst_total
```

```python
analyst_total = pd.merge(analyst_total,analyst_XR,how="outer",on="Case ID")
analyst_total
```

```python
analyst_total = pd.merge(analyst_total,analyst_flup,how="outer",on="Case ID")
analyst_total
```

```python
analyst_total = pd.merge(analyst_total,analyst_app,how="outer",on="Case ID")
analyst_total
```

Verifying column names

```python
analyst_total.fillna(value=0,inplace=True)
analyst_total.columns
```

Finding total duration

```python
column1 = ['PA duration', 'IX duration', 'PR duration', 'XR duration',
       'FLUP duration', 'Approval duration']
analyst_total[column1].astype('int64')
analyst_total['Total Duration'] = analyst_total[column1].sum(axis=1)
analyst_total
```

Formatting of Time Duration into proper format

```python
column2 = ['PA duration', 'IX duration', 'PR duration', 'XR duration',
       'FLUP duration', 'Approval duration', 'Total Duration']
analyst_total_hours = analyst_total
analyst_total_hours

for i in analyst_total_hours.index:
    for col in column2:
        duration = analyst_total_hours.loc[i,col]
        hours = int(duration // 60)
        minutes1 = int(duration % 60)
        seconds = int(((duration % 60) % 1)*60)
        time_string = "{}:{}:{}".format(hours, minutes1, seconds)
        analyst_total_hours.loc[i,col] = time_string
```
