Concatenating the needed reports only

```python
all_surveys_list = pd.concat([TMH,NCG],ignore_index=True).reset_index()
all_surveys_list.drop('index', axis=1, inplace=True)

all_surveys_list.info()
```

Filtering out completed surveys

```python
all_comp_surveys_list = all_surveys_list[all_surveys_list['STATUS']=='Completed']
```

```python
all_comp_surveys_list.reset_index()
all_comp_surveys_list.info()
```

```python
all_comp_surveys_list.reset_index(drop=True)
all_comp_surveys_list.info()
```

Merging columns and renaming alonng with dropping unneeded columns for cleaner data

```python
for i in all_comp_surveys_list.index:
    all_comp_surveys_list.loc[i,"SENT TIME"] = all_comp_surveys_list.loc[i,"SENT"]+", "+all_comp_surveys_list.loc[i,"Unnamed: 7"]
```

```python
for i in all_comp_surveys_list.index:
    all_comp_surveys_list.loc[i,"RESPONDED AT"] = all_comp_surveys_list.loc[i,"RESPONDED"]+", "+all_comp_surveys_list.loc[i,"Unnamed: 9"]
    
all_comp_surveys_list.head()

all_comp_surveys_list.drop(["SENT","Unnamed: 7", "RESPONDED", "Unnamed: 9", "SOURCE", "DEVICE"], axis=1,inplace=True)
```

Formatting columns with time durations into proper format and saving to CSV

```python
all_comp_surveys_list['DURATION'] = pd.to_datetime(all_comp_surveys_list.DURATION, unit='s').dt.strftime('%H:%M:%S')
all_comp_surveys_list['TURNAROUND'] = pd.to_datetime(all_comp_surveys_list.TURNAROUND, unit='s').dt.strftime('%H:%M:%S')

all_comp_surveys_list.to_csv('All Expert Responded Surveys.csv')
```

Creating speciality responses per day report using the completed report

```python
test = all_comp_surveys_list

test = test.set_index(test.columns.drop('DMG',1).tolist()).DMG.str.split(',', expand=True).stack().reset_index().rename(columns={0:'DMG'}).loc[:, test.columns]

test = test.join(test['RESPONDED AT'].str.split(',', 2, expand=True).rename(columns={0:'Date', 1:'Day', 2: 'Time'}))

specialty_by_day = test.groupby(['DMG','Day']).size().reset_index(name = 'Number of Responses')

specialty_by_day.to_csv("Specialty Responses per days of week.csv")
```

Splitting up rows with comma separated DMG values into multiple rows

```python
turnaround_test = all_surveys_list[all_surveys_list['STATUS']=='Completed']
turnaround_test = turnaround_test.set_index(turnaround_test.columns.drop('DMG',1).tolist()).DMG.str.split(',', expand=True).stack().reset_index().rename(columns={0:'DMG'}).loc[:, turnaround_test.columns]
```

Finding Average Turnaroud time for each DMG

```python
turnaround_avg = turnaround_test.groupby('DMG')['TURNAROUND','DURATION'].mean().reset_index().rename(columns={'TURNAROUND': 'AVERAGE_TURNAROUND', 'DURATION': 'AVERAGE_DURATION'})


```

```python
turnaround_avg
```

Introduction of proper formatting

```python
turnaround_avg['AVERAGE_DURATION'] = pd.to_datetime(turnaround_avg.AVERAGE_DURATION, unit='s').dt.strftime('%H:%M:%S')
turnaround_avg['AVERAGE_TURNAROUND'] = pd.to_datetime(turnaround_avg.AVERAGE_TURNAROUND, unit='s').dt.strftime('%H:%M:%S')
```

```python
turnaround_avg.to_csv('Average Turnaround per specialty.csv')
```
